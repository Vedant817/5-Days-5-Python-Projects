```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the Data
df = pd.read_csv("./diabetese.tsv", sep="\t")
df.head()
```

```python
df.info()
```

```python
# Lets find the Average Age of diabetic patients
mean_age = df['AGE'].mean()
mean_age
```

```python
# Variance: How is the value different from the mean value.
var_Age = df['AGE'].var()
var_Age
```

Standard Deviation helps in finding out how much data is spread, which means that how much data is varying with respect to the mean. Higher the value means data's values is well distributed.

```python
# Standard Deviation
std_Age = df['AGE'].std()
print(f"Standard Deviation is {std_Age}")
```

```python
# Box Plot
plt.figure(figsize=(2,6))
plt.boxplot(df['AGE'], showmeans=True)
plt.grid(color= 'gray', linestyle='dotted')
plt.show()
```

```python
# Boc Plot gor BMI
plt.figure(figsize=(6,2))
plt.boxplot(df['BMI'], vert=False, showmeans=True)
plt.grid(color='gray', linestyle='dotted')
plt.show()
```

```python
# Boc Plot gor BP
plt.figure(figsize=(6,2))
plt.boxplot(df['BP'], vert=False, showmeans=True)
plt.grid(color='gray', linestyle='dotted')
plt.show()
```
